```
#dataset property
import pandas as pd
prp = pd.read_csv('property.csv')
```

```
# menghapus baris dan kolom yang kosong
prp.isnull().sum()
```

```
prp['PID']
```

```
prp['PID'].fillna(100008000.0, inplace=True)
```

```
prp['PID']
```

```
prp['ST_NUM']
```

```
prp['ST_NUM'].fillna(197.0, inplace=True)
```

```
prp['OWN_OCCUPIED']
```

```
prp['OWN_OCCUPIED'].fillna(value='N', inplace=True)
```

```
#merubah value pada OWN_OCCUPIED yang aslinya 12 menjadi N

prp['OWN_OCCUPIED'] = prp['OWN_OCCUPIED'].replace('12', 'N')
```

```
#mengganti value NaN pada kolom NUM_BEDROOMS dengan angka 2

prp['NUM_BEDROOMS'].fillna(2, inplace=True)
```

```
#merubah value pada OWN_OCCUPIED yang aslinya na menjadi 3

prp['NUM_BEDROOMS'] = prp['NUM_BEDROOMS'].replace('na', '3')
```

```
#mengganti value NaN dan HURLEY pada kolom NUM_BATH dengan angka 1,5

prp['NUM_BATH'].fillna(1.5, inplace=True)
prp['NUM_BATH'] = prp['NUM_BATH'].replace('HURLEY', '1.5')
```

```
#mengganti value -- dan NaN pada kolom SQ_FT dengan angka 1800 dan 850

prp['SQ_FT'].fillna(1800, inplace=True)
prp['SQ_FT'] = prp['SQ_FT'].replace('--', '850')
```

```
prp
```

```
prp['ST_NUM'].fillna(method='bfill', inplace=True)
```

```
prp['ST_NUM']
```

```
median = prp['ST_NUM'].median()
```

```
median
```

```
# menghapus baris dan kolom yang kosong
prp.dropna()
```
